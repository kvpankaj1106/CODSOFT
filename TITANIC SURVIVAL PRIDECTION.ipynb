```python
# change the datatype 
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
```

```python
df["Age"] = df.Age.astype(int)
df["Fare"] = df.Fare.astype(int)
```

# DATA VISUALIZATION

```python

fig = plt.figure(figsize =(15, 5))
plt.hist(x = [df[df['Survived']==1]['Age'], df[df['Survived']==0]['Age']],stacked=True, color = ['g','r'],label = ['Survived','Not survived'],rwidth=.9)
plt.title('Age Histogram with Survival')
plt.xlabel('Age')
plt.ylabel('No of passengers')
plt.legend()
```

```python
df.columns
```

```python
sns.countplot(x=df['Survived'],data=df,hue='Survived')
```

```python
df=df["Sex"].value_counts().plot(kind="pie", autopct="%.2f%%")
```

```python
_, ax = plt.subplots(figsize = (10, 7))
sns.histplot(data = df, x = "Age", hue = "Survived", multiple = "stack", kde = True, palette=["red", "green"], ax = ax)
ax.set_title("Histogram for Age with Survival")
ax.set_xlabel("Age")
ax.set_ylabel("Number of passengers")
ax.legend(title = "Legends", labels = ["Survived", "Not survived"])
plt.plot()
```
